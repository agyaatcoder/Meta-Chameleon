### Multimodal Input

```python
import urllib.request

def download_image(url, save_as):
    urllib.request.urlretrieve(url, save_as)

image_url = 'https://res.cloudinary.com/demo/image/upload/v1312461204/sample.jpg'
save_as = 'image.jpg'

download_image(image_url, save_as)
```

```python
from PIL import Image
```

```python
Image.open('image.jpg')
```

```python
tokens = model.generate(
    prompt_ui=[
        {"type": "image", "value": "file:image.jpg"},
        {"type": "text", "value": "What do you see?"},
        {"type": "sentinel", "value": "<END-OF-TURN>"},
    ]
)
print(model.decode_text(tokens)[0])
```
